```python
import os
from functools import partial
import datasets
from _utils.electra_dataprocessor import ELECTRADataProcessor
from hugdatafast import HF_Datasets
from fastai.text.all import TensorText, noop

from transformers import ElectraConfig, ElectraTokenizerFast, ElectraForMaskedLM, ElectraForPreTraining
```

```python
hf_tokenizer = ElectraTokenizerFast.from_pretrained(f"google/electra-small-generator")
```

```python
ELECTRAProcessor = partial(ELECTRADataProcessor, hf_tokenizer=hf_tokenizer, max_length=128)
```

```python
# Wikipedia
dsets = []
print('load/download wiki dataset')
wiki = datasets.load_dataset('wikipedia', '20200501.en', cache_dir='./datasets')['train']
```

```python
wiki.select(range(128)).ma
```

```python
e_wiki = ELECTRAProcessor(wiki.select(range(129))).map(num_proc=1)
```

```python
def hoge(texts):
    print(texts.keys())
    print(len(texts['title']),texts['title'])

wiki.select(range(1290)).map(hoge, batched=True)
```

```python
for a in e_wiki:
    print(a.keys())
```

```python
merged_dsets = {'train': datasets.concatenate_datasets(dsets)}

hf_dsets = HF_Datasets(
        merged_dsets,
        cols={'input_ids': TensorText, 'sentA_length': noop},
        hf_toker=hf_tokenizer, n_inp=2)

dls = hf_dsets.dataloaders(
        bs=128,
        num_workers=3,
        pin_memory=False,
        shuffle_train=True,
        srtkey_fc=False,
        cache_dir='./datasets/electra_dataloader',
        cache_name='dl_{split}.json')
```
